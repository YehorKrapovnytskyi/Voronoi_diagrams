```python
import torch
from torch.utils.data import DataLoader
import sys
import os
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
sys.path.append(os.path.abspath('../src/'))
from dataset.VoronoiDataset import VoronoiDataset
```

```python
dataset = VoronoiDataset("../voronoi_data/voronoi_data.h5")
```

```python
random.seed(6)
random_images_nums = random.sample(range(0, len(dataset)), 5)
random_images_nums
```

```python
fig, ax = plt.subplots(5, 1, figsize=(10, 10), dpi=500)
for idx, i in enumerate(random_images_nums):
    image = dataset[i]["image"]
    print(dataset[i]["points"][0])
    image_permuted = image.permute(1, 2, 0)
    ax[idx].imshow(image_permuted, cmap="gray")
    ax[idx].axis('off') 

plt.tight_layout()
```
